```python
import pandas as pd
import altair as alt


url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/licenses_fall2022.csv"
data = pd.read_csv(url)

data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
columns_to_drop = ['Title', 'Prefix', 'Suffix', 'BusinessDBA','Middle']  # Replace with actual column names
data_without_columns = data.drop(columns=columns_to_drop)
```

```python
cleaned_data = data_without_columns.dropna()

print(f"Data shape after cleaning: {cleaned_data.shape}")
cleaned_data.head()
```

```python
cleaned_data.shape
```

```python
cleaned_data.dtypes
```

```python
aggregated_data = cleaned_data.groupby(['License Type', 'Business']).size().reset_index(name='Count')

interactive_bar_chart = alt.Chart(aggregated_data).mark_bar().encode(
    x=alt.X('Count:Q', title='Count of Licenses'),
    y=alt.Y('License Type:N', title='License Type', sort='-x'),
    color=alt.Color('Business:N', legend=None),
    tooltip=['License Type', 'Business', 'Count']
).properties(
    title="Interactive Bar Chart: License Type vs Business",
    width=600,
    height=400
).interactive()  

interactive_bar_chart
```

```python

licenses_discipline = cleaned_data.groupby('Ever Disciplined').size().reset_index(name='License Count')

bar_chart = alt.Chart(licenses_discipline).mark_bar().encode(
    x=alt.X('Ever Disciplined:N', title='Ever Disciplined'),
    y=alt.Y('License Count:Q', title='Number of Licenses'),
    color=alt.Color('Ever Disciplined:N', title='Ever Disciplined', scale=alt.Scale(scheme='blues')),
    tooltip=['Ever Disciplined', 'License Count']
).properties(
    title="Licenses by Ever Disciplined Status",
    width=600,
    height=400
)

bar_chart
```

```python
bar_chart.save('assets/bar_chart.html')
interactive_bar_chart.save('assets/interactive_bar_chart.html')
```
